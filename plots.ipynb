```python
from datetime import datetime
import pandas as pd
import matplotlib as plot
import seaborn as sns
import numpy as np
from pytz import timezone
import pytz
%matplotlib inline
```

```python
def to_pst(dt, dateformat='%Y-%m-%d %I%p'):
    return dt.replace(tzinfo=pytz.utc).astimezone(timezone('US/Pacific')).strftime(dateformat)
```

```python
headers = ['datetime', 'temp', 'temp-avg', 'humid', 'humid-avg']
df = pd.read_csv("data.csv", header=None, names=headers, parse_dates=True, index_col='datetime')
coolheaders = ['datetime', 'seconds', 'target', 'actual', 'starting']
dfcool = pd.read_csv("lcool.csv", header=None, names=coolheaders, parse_dates=True, index_col='datetime')
df.head(), dfcool.head()
```

```python
df['temp'] = df['temp'].resample('1S', how='mean')
df['temp-avg'] = df['temp-avg'].resample('1S', how='mean')
df['humid'] = df['humid'].resample('1S', how='mean')
df['humid-avg'] = df['humid-avg'].resample('1S', how='mean')
```

```python
start_date = '2015-08-10T01:00:00'
start_date = '2015-08-25T00:00:00'
start_date = '2015-08-28T02:00:00'
start_date = '2015-09-01T00:00:00'

recent = df[df.index > start_date]
resampled = recent.resample('30Min', how='mean')
recent.shape, resampled.shape
```

```python
def plot_with_avgline(data, span=25):
    try:
        data.plot()
    except: pass
    try:
        pd.ewma(data, span=span, adjust=True).plot(style='k')
    except: pass
```

```python
# plot_with_avgline(recent['temp-avg'], span=60*5)
plot_with_avgline(resampled['temp-avg'], span=25)
```

```python
# plot_with_avgline(recent['humid-avg'], span=60*5)
plot_with_avgline(resampled['humid-avg'], span=25)
```

```python
def hist(data, datatype, bins=20):
    ax = data.hist(bins=bins)
    text = '$\mu={:.2f}$\n$\sigma={:.2f}$'.format(data.mean(), data.std())
    props = dict(facecolor='white', alpha=0.75)
    plot.pyplot.text(0.05, 0.95, text, transform=ax.transAxes, fontsize=14, verticalalignment='top', bbox=props)
    plot.pyplot.xlabel('{} between {} and {}'.format(datatype, to_pst(data.index.min()), to_pst(data.index.max())))
    ax.set_yticklabels([])
```

```python
hist(recent['temp-avg'], 'degrees fahrenheit')
```

```python
hist(recent['humid-avg'], 'humidity')
```
