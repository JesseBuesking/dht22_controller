```python
from datetime import datetime
import pandas as pd
import matplotlib as plot
import seaborn as sns
import numpy as np
from pytz import timezone
import pytz
```

```python
def to_pst(dt, dateformat='%Y-%m-%d %I%p'):
    return dt.replace(tzinfo=pytz.utc).astimezone(timezone('US/Pacific')).strftime(dateformat)
```

```python
headers = ['datetime', 'temp', 'temp-avg', 'humid', 'humid-avg']
df = pd.read_csv("data.csv", header=None, names=headers, parse_dates=True, index_col='datetime')
df.head()
```

```python
temp = df['temp-avg'].resample('1S', how='mean')
temp.head()
```

```python
# sns.tsplot(temp)
temp.plot()
```

```python
# temp2 = temp[temp.shape[0] - temp.shape[0]:]
temp2 = temp[temp.index > '2015-08-10T01:00:00']
temp2.shape
```

```python
temp2.resample('30Min', how='mean').plot()
pd.ewma(temp2.resample('30Min', how='mean'), span=25, adjust=True).plot(style='k')
```

```python
pd.tools.plotting.autocorrelation_plot(temp2.resample('30Min', how='mean'))
```

```python
pd.tools.plotting.bootstrap_plot(temp2.resample('1Min', how='mean'))
```

```python
ax = temp2.hist(bins=20)
text = '$\mu={:.2f}$\n$\sigma={:.2f}$'.format(temp2.mean(), temp2.std())
props = dict(facecolor='white', alpha=0.75)
plot.pyplot.text(0.05, 0.95, text, transform=ax.transAxes, fontsize=14, verticalalignment='top', bbox=props)
plot.pyplot.xlabel('degrees fahrenheit between {} and {}'.format(to_pst(temp2.index.min()), to_pst(temp2.index.max())))
ax.set_yticklabels([])
```
